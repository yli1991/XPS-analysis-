```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/Users/liyi/Desktop/online class/Python/XPS data/1TMA-C.csv')
```

```python
df.drop('Unnamed: 9', axis = 1, inplace = True)
```

```python
df.drop('Residuals', axis = 1, inplace = True)
```

```python
df.drop('Unnamed: 1', axis = 1, inplace = True)
```

```python
df.columns= ['BE', 'C1', 'C2','C3','C4', 'Background','Envelop', 'Nan']
df.drop('Nan', axis = 1, inplace = True)
df
```

```python
fig1 = px.line(df, x = df['BE'], y =df['C1'])

fig1.add_scatter( x = df['BE'], y =df['C2'] )

fig1.add_scatter( x = df['BE'], y =df['C3'] )

fig1.add_scatter( x = df['BE'], y =df['C4'] )

fig1.add_scatter( x = df['BE'], y =df['Background'] )

fig1.add_scatter( x = df['BE'], y =df['Envelop'] )

for i in range (len(str(df.columns))-1):
    fig1.for_each_trace(lambda x: x.update(name= str(df.columns[i])))

fig1.show()
#fig1.update_traces(name='Envelop', showlegend = True)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter( x = df['BE'], y =df['C1'], mode = 'lines+markers', name = 'C1'))
fig.add_trace(go.Scatter( x = df['BE'], y =df['C2'], mode = 'lines+markers', name = 'C2'))
fig.add_trace(go.Scatter( x = df['BE'], y =df['C3'], mode = 'lines+markers', name = 'C3'))
fig.add_trace(go.Scatter( x = df['BE'], y =df['C4'], mode = 'lines+markers', name = 'C4'))
fig.add_trace(go.Scatter( x = df['BE'], y =df['Background'], mode = 'lines+markers', name = 'Background'))
fig.add_trace(go.Scatter( x = df['BE'], y =df['Envelop'], mode = 'lines+markers', name = 'Envelop'))

fig.update_layout( title = 'Normlized Fitted C XPS spectrum',
                   xaxis_title = 'Binding Energy',
                   yaxis_title = 'Counts',
                   legend_title_text = 'Fitted C')

fig.show()
```
